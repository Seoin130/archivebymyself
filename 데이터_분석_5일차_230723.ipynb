```python
import pandas as pd
```

```python
ns_df=pd.read_csv('ns_202104.csv', low_memory=False)
ns_df.head()
```

unnamed: 13 열은 csv 각 라인 끝에 콤마가 있어서 판다스가 자동으로 추가한 것
삭제해 줘야 함

```python
ns_book=ns_df.loc[:, '번호':'등록일자']
ns_book.head()
```

불리언 배열을 사용하여 중간에 있는 열을 제외하고 표를 만들 수 있음
사용할 열을 불리언 배열로 나타내려면 true, false와 같은 불리언 값을 일일이 나열하여 작성해야 함

```python
selected_columns=ns_df.columns!='Unnamed: 13'
ns_book=ns_df.loc[:, selected_columns]
```

```python
ns_book.head()
```

```python
selected_columns=ns_df.columns!='부가기호'
ns_book=ns_df.loc[:, selected_columns]
ns_book.head()
```

drop() 메서드: 행이나 열을 삭제하는 함수

```python
ns_book=ns_df.drop('Unnamed: 13', axis=1)
ns_book.head()
```

axis를 1로 설정하며 열을 삭제 함
axis 이전에 삭제 하려는 열 이름을 넣어야 함

```python
ns_book=ns_df.drop(['부가기호', 'Unnamed: 13'], axis=1)
ns_book.head()
```

여러 개의 열을 삭제하려면 리스트 형식으로 넣어서 삭제

```python
ns_book.drop('주제분류번호', axis=1, inplace=True)
ns_book.head()
```

여태까지의 삭제 방법은 ns_df의 정보에서 drop한 것을 ns_book에 저장했는데, 이제는 ns_df 자체에서 삭제할 수 있도록 함

inplace를 True로 지정하면 현재 선택한 데이터 프레임을 바로 수정할 수 있음

```python
ns_book=ns_df.dropna(axis=1)
ns_book.head()
```

dropna() 메서드는 NaN이 포함된 하나 이시ㅏㅇ의 행이나 열을 삭제 함

```python
ns_book=ns_df.dropna(axis=1, how='all')
ns_book.head()
```

모든 값이 NaN인 열을 삭제하려면 how 매개 변수를 all로 지정하면 됨
