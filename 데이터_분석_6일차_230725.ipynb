```python
sum(ns_book.duplicated())
```

duplicated() 메서드를 통하여 정보 내 중복되는 것이 있는 지 확인 함. sum()은 true를 1로 인식하기 때문에 sum()의 값으로 중복되는 값이 몇 개나 있는 지 확인 가능
duplicated()는 모든 열을 기준으로 중복된 행을 찾음

```python
sum(ns_book.duplicated(subset=['도서명', '저자', 'ISBN']))
```

subset 매개변수에 기준 열을 나열함

```python
dup_rows=ns_book.duplicated(subset=['도서명', '저자', 'ISBN'], keep=False)
ns_book3=ns_book[dup_rows]
ns_book3.head()
```

keep 매개변수를 False로 지정하여 중복된 모든 행을 True로 표시함
